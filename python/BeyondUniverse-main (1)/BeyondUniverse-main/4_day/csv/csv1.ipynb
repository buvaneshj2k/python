Reading and writing CSV (Comma Separated Values) files is a common task in data processing and analysis with Python. Here's how you can work with CSV files using the built-in csv module:
Reading CSV Files:

    Using csv.reader:
        Use csv.reader to read rows from a CSV file.

```python
import csv

# Open the CSV file in read mode
with open('tips.csv', mode='r') as file:
    # Create a CSV reader object
    csv_reader = csv.reader(file)
    
    # Iterate over each row in the CSV file
    for row in csv_reader:
        print(row)
```

    Using DictReader:
        csv.DictReader reads each row as an OrderedDict (for Python 3.6+) or a regular dictionary (for earlier versions).

```python
import csv

# Open the CSV file in read mode
with open('tips.csv', mode='r') as file:
    # Create a CSV DictReader object
    csv_reader = csv.DictReader(file)
    
    # Iterate over each row (each row is a dictionary)
    for row in csv_reader:
        print(row)
        # Access specific columns using keys, e.g., row['column_name']
```

Writing CSV Files:

    Using csv.writer:
        Use csv.writer to write rows to a CSV file.

```python
import csv

# Sample data
data = [
    ['Name', 'Age', 'City'],
    ['Alice', 25, 'New York'],
    ['Bob', 30, 'Los Angeles'],
    ['Charlie', 22, 'Chicago']
]

# Open the CSV file in write mode
with open('output.csv', mode='w', newline='') as file:
    # Create a CSV writer object
    csv_writer = csv.writer(file)
    
    # Write data to the CSV file
    for row in data:
        csv_writer.writerow(row)

print('CSV file written successfully.')
```

    Using DictWriter:
        csv.DictWriter writes dictionaries to CSV files.

```python
import csv

# Sample data as dictionaries
data = [
    {'Name': 'Alice', 'Age': 25, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 30, 'City': 'Los Angeles'},
    {'Name': 'Charlie', 'Age': 22, 'City': 'Chicago'}
]

# Define CSV fieldnames (column names)
fieldnames = ['Name', 'Age', 'City']

# Open the CSV file in write mode
with open('output.csv', mode='w', newline='') as file:
    # Create a CSV DictWriter object
    csv_writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    # Write CSV header
    csv_writer.writeheader()
    
    # Write data to the CSV file
    for row in data:
        csv_writer.writerow(row)

print('CSV file written successfully.')
```
